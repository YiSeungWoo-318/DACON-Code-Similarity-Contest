```python
import os
import pandas as pd
import numpy as np
from glob import glob
from tqdm import tqdm
```

```python
test = pd.read_csv("D:/code_classification/test.csv")

temp = pd.concat([test.code1, test.code2], ignore_index=True)

re_test = pd.DataFrame(data={"code": temp})

re_test.drop_duplicates("code", inplace=True)

re_test.reset_index(drop=True, inplace=True)

re_test["id"] = re_test.index

code2id = {}
for code, ids in zip(re_test["code"], re_test["id"]):
    code2id[code] = ids

test["id1"] = test["code1"].apply(lambda x: code2id[x])
test["id2"] = test["code2"].apply(lambda x: code2id[x])
```
